```python
import pandas as pd
```

```python
edgesdf = pd.read_csv('Data/edges.tsv', sep = '\t')
nodesdf = pd.read_csv('Data/nodes.tsv', sep ='\t')
```

```python
compounds = nodesdf[nodesdf["kind"] == "Compound"]['id']
```

```python
ctd = edgesdf[
    edgesdf["source"].str.startswith("Compound") & edgesdf["target"].str.startswith("Disease")
]
```

```python
ctd = ctd['source'].unique()
ctd
```

```python
cwd = compounds[~compounds.isin(ctd)]
```

```python
cwd.count()
```

```python
anatomy = nodesdf[nodesdf["kind"] == "Anatomy"]['id']
```

```python
dta = edgesdf[
    (edgesdf["source"].str.startswith("Disease::")) &
    (edgesdf["target"].isin(anatomy))
]

awd = dta["target"].unique()

awd = anatomy[~anatomy.isin(awd)]

awd.tolist()
awd
```

```python
cta = edgesdf[
    (edgesdf["source"].str.startswith("Compound::")) &
    (edgesdf["target"].isin(awd))
]

cawd = cta["source"].unique()

cawd.tolist()
cawd
```

```python
AdG = edgesdf[edgesdf["metaedge"] == 'AdG']
AuG = edgesdf[edgesdf["metaedge"] == 'AuG']
CdG = edgesdf[edgesdf["metaedge"] == 'CdG']
CuG = edgesdf[edgesdf["metaedge"] == 'CuG']
```

```python
AdG_CuG_merged = pd.merge(AdG, CuG, on="target", suffixes=("_AdG", "_CuG"), how='inner')
unique_compounds1 = AdG_CuG_merged["source_CuG"].unique()
```

```python
AuG_CdG_merged = pd.merge(AuG, CdG, on="target", suffixes=("_AuG", "_CdG"), how='inner')
unique_compounds2 = AuG_CdG_merged["source_CdG"].unique()
```

```python
combined_compounds = set(unique_compounds1) | set(unique_compounds2)
combined_compounds
```

```python
cwd_set = set(cwd.values)
```

```python
filtered_compounds = [compound for compound in combined_compounds if compound in cwd_set]
len(filtered_compounds)
```
